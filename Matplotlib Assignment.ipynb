```python
#Q.2 
# A scatterplot shows the relationship between two quantitative variables measured for the same individuals 
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("plot")
plt.show()
```

```python
#Q.4
# The bar() function is used to create a bar plot that is bounded with a rectangle depending on the given parameters of the function
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```
